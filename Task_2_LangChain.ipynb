```python
from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline

pipe = pipeline("text-generation", model="gpt2")

llm = HuggingFacePipeline(pipeline=pipe)
```

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate.from_template(
    "Explain {topic} in simple terms"
)

result = (prompt | llm).invoke({"topic": "LangChain"})
print(result)
```

```python
chain = prompt | llm

print(chain.invoke({"topic": "Artificial Intelligence"}))
```

```python
# Simple manual memory (safe & no errors)

chat_history = []

def chat(input_text):
    chat_history.append(f"User: {input_text}")

    context = "\n".join(chat_history)

    response = llm.invoke(context + "\nAI:")

    chat_history.append(f"AI: {response}")

    return response

print(chat("Hi, my name is Shital"))
print(chat("What is my name?"))
```

```python
from langchain.tools import tool

@tool
def calculator(expression: str):
    """Evaluates a mathematical expression."""
    return str(eval(expression))

print(calculator.run("25 * 4"))
```

```python
# Simple agent simulation (avoids LangChain agent errors)

query = "25 * 4"

print("Agent Output:")
print(calculator.run(query))
```

```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import FakeEmbeddings

embedding = FakeEmbeddings(size=10)

texts = ["LangChain is powerful", "It helps build LLM apps"]

db = Chroma.from_texts(texts, embedding)

results = db.similarity_search("What is LangChain?")
print(results)
```
